```python
from google.colab import drive
drive.mount('/content/drive')

cpp_code = """
#include <iostream>

using namespace std;

int main() {
    int choice;
    double time, result;
    char confirm;

    while (true) {
        // Display the menu

        cout << "Time Conversion Menu" << endl;
        cout << "1. Convert seconds to hours" << endl;
        cout << "2. Convert seconds to minutes" << endl;
        cout << "3. Convert minutes to hours" << endl;
        cout << "4. Convert hours to seconds" << endl;
        cout << "5. Exit" << endl;
        cout << "Enter your choice: ";

        // Check if the input is an integer
        if (!(cin >> choice)) {
            cout << "Invalid input. Please enter an integer." << endl;
            cin.clear();
            continue;
        }

        switch (choice) {
            case 1: // Convert seconds to hours
                cout << "Enter time in seconds: ";
                cin >> time;
                if (time < 0) {
                    cout << "Invalid input. Time cannot be negative." << endl;
                    continue;
                }
                result = time / 3600.0;
                cout << time << " seconds is equal to " << result << " hours" << endl;
                break;
            case 2: // Convert seconds to minutes
                cout << "Enter time in seconds: ";
                cin >> time;
                if (time < 0) {
                    cout << "Invalid input. Time cannot be negative." << endl;
                    continue;
                }
                result = time / 60.0;
                cout << time << " seconds is equal to " << result << " minutes" << endl;
                break;
            case 3: // Convert minutes to hours
                cout << "Enter time in minutes: ";
                cin >> time;
                if (time < 0) {
                    cout << "Invalid input. Time cannot be negative." << endl;
                    continue;
                }
                result = time / 60.0;
                cout << time << " minutes is equal to " << result << " hours" << endl;
                break;
            case 4: // Convert hours to seconds
                cout << "Enter time in hours: ";
                cin >> time;
                if (time < 0) {
                    cout << "Invalid input. Time cannot be negative." << endl;
                    continue;
                }
                result = time * 3600.0;
                cout << time << " hours is equal to " << result << " seconds" << endl;
                break;
            case 5: // Exit the program with confirmation
                cout << "Are you sure you want to exit? (y/n): ";
                cin >> confirm;
                if (confirm == 'y' || confirm == 'Y') {
                    cout << "Exiting the program. Goodbye!" << endl;
                    return 0;
                } else {
                    cout << "Returning to the menu." << endl;
                }
                break;
            default:
                cout << "Invalid choice. Please enter a number between 1 and 5." << endl;
        }
    }

    return 0;
}
"""
 #path to save in google drive
cpp_file_path = "/content/drive/MyDrive/_calculator.cpp"

with open(cpp_file_path, "w") as file:
    file.write(cpp_code)

    print(f"C++ code saved to {cpp_file_path}")
```
